```python
import numpy as np
import mne
import matplotlib.pyplot as plt

def read_biologic_eeg_full(filename, sfreq=256):
    """Read BioLogic EEG file without truncation"""
    # Full channel configuration (42 channels)
    ch_names = [
        'PG1', 'PG2', 'FP1', 'FP2', 'F7', 'F3', 'FZ', 'F4', 'F8',
        'A1', 'T3', 'C3', 'CZ', 'C4', 'T4', 'A2', 'T5', 'P3', 'PZ',
        'P4', 'T6', 'O1', 'OZ', 'O2', 'X1', 'Y1', 'X2', 'Y2', 'X3',
        'Y3', 'X4', 'Y4', 'X5', 'Y5', 'X6', 'Y6', 'X7', 'Y7', 'X8',
        'Y8', 'PHOT', 'NoRef'
    ]
    
    # Read binary data
    with open(filename, 'rb') as f:
        data = np.fromfile(f, dtype=np.float32)
    
    # Calculate dimensions without truncation
    n_channels = len(ch_names)
    n_samples = len(data) // n_channels
    remainder = len(data) % n_channels
    
    if remainder != 0:
        print(f"Note: File contains {len(data)} samples ({remainder} extra samples)")
    
    # Pad with zeros to make complete channels
    if remainder > 0:
        pad_size = n_channels - remainder
        data = np.concatenate([data, np.zeros(pad_size)])
        n_samples += 1
    
    # Reshape with Fortran order (column-major)
    data = data.reshape((n_channels, n_samples), order='F')
    
    # Create MNE object
    ch_types = ['stim']*2 + ['eeg']*22 + ['misc']*18  # Adjust as needed
    info = mne.create_info(ch_names, sfreq, ch_types)
    raw = mne.io.RawArray(data, info)
    
    return raw

# Load your file
raw = read_biologic_eeg_full('/Users/yann/Documents/NUÑE 07_12_2011_____ 2025MAY10 091626 C1 B.EEG', sfreq=256)
```

```python
import pandas as pd
import numpy as np

def create_biologic_dataframe(raw):
    """
    Creates a comprehensive DataFrame for BioLogic EEG data with voltage and channel metadata.
    
    Args:
        raw: MNE Raw object from read_biologic_eeg_full()
    
    Returns:
        tuple: (voltage_df, channel_df, recording_metadata)
    """
    # 1. Voltage Data DataFrame (time × channels)
    voltage_data = raw.get_data() * 1e6  # Convert to µV
    voltage_df = pd.DataFrame(
        data=voltage_data.T,
        columns=raw.ch_names,
        index=pd.Index(raw.times, name='time_seconds')
    )
    
    # 2. Channel Information DataFrame
    channel_df = pd.DataFrame({
        'channel_name': raw.ch_names,
        'channel_type': raw.get_channel_types(),
        'unit': 'µV',
        'sampling_rate': raw.info['sfreq'],
        'in_head': [False]*len(raw.ch_names)  # BioLogic typically doesn't have digitization
    })
    
    # 3. Recording Metadata
    recording_metadata = {
        'equipment': 'BioLogic',
        'duration': '00:06:57',
        'sampling_frequency': 256.00,
        'time_points': 106646,
        'filters': {
            'highpass': 0.00,
            'lowpass': 128.00
        },
        'channels_summary': {
            'EEG': 22,
            'misc': 18,
            'Stimulus': 2
        }
    }
    
    return voltage_df, channel_df, recording_metadata

# Create the DataFrames
voltage_df, channel_df, meta = create_biologic_dataframe(raw)

# Display information
print("Voltage DataFrame:")
print(voltage_df.head())
print(f"\nShape: {voltage_df.shape} (samples × channels)")

print("\nChannel Information:")
print(channel_df.head(25))  # Show first 25 channels
```

```python
cropped_df = voltage_df.iloc[:, 2:24]  
cropped_df
```

```python
fm=256 #frecuencia de muestreo
ti=5 #tiempo de inicio, debe ser número entero
tf=6 #tiempo final, debe ser número entero
pt=(tf-ti)*fm#numero de puntos
t1=ti*fm #nos ayuda a dar la posición de la fila para el tiempo inicial
t2=tf*fm #nos ayuda a dar la posición de la fila para el tiempo final
nombres=list(cropped_df) #almacena los nombres de los encabezados en una lista
t=np.linspace(ti,tf,pt) 
for i in range (0,21):
    plt.figure(figsize=(15,12))
    plt.subplot(21,1,i+1) #(filas, columnas, i)
    plt.plot(t,np.array(cropped_df)[t1:t2,i]) #t tiempo, i posición
    plt.ylabel(nombres[i])#itera sobre el nombre de las columnas para dar título a los gráficos
    if i == 21:  # Si es el último subplot
        plt.xlabel('Tiempo [s]')  # Agregar la etiqueta al eje x
    plt.show()
# Etiquetar tiempo en la última línea del subplot
```
