## 5.1 Introducción

El EEG se analiza principalmente en bandas de frecuencia bajas, entre 0.5 y 40 Hz,  
donde se ubican ritmos como alfa (8–13 Hz) y beta (13–30 Hz) relacionados con actividad cognitiva y motora (Allen & Rabiner, 1977).  
El EMG concentra su energía en un rango más alto (20–450 Hz), propio de la descarga rápida de fibras musculares superficiales (Merletti & Hermens, 2000).  

Comparar ambas señales permite entender cómo los mismos métodos —filtrado, segmentación, métricas de energía—  
se aplican de manera diferente según el fenómeno fisiológico de origen.  
En palabras de Addison (2017), la bioseñal determina el método, y elegir correctamente el análisis es lo que permite interpretar la señal con sentido.

## 5.2.1 Carga de datos

Los datos se cargan desde la carpeta local donde se almacenan todos los registros de este proyecto.  
El EEG está en formato `.csv` y el EMG en `.wav`.  

En el caso del EMG, usamos y adaptamos código de Erin McKiernan (2025), quien desarrolló materiales abiertos de enseñanza en procesamiento de señales biomédicas.  
Mostrar el registro crudo es un primer paso esencial: antes de filtrar o calcular métricas, conviene visualizar cómo luce la señal en bruto y con qué parámetros fue adquirida.


```python
EEG_CSV = '/Users/yann/Documents/Tesis/Datos/EEG/H24_OC_OA.csv'
EMG_WAV = '/Users/yann/Documents/Tesis/Datos/EMG/S10_EMG_calf_intermittent.wav'

# Cargar EEG
df_eeg = pd.read_csv(EEG_CSV)
print("EEG cargado:", df_eeg.shape)

# Cargar EMG crudo (código adaptado de McKiernan, 2025)
record = wave.open(EMG_WAV, 'r')
numChannels = record.getnchannels()
N = record.getnframes()
sampleRate = record.getframerate()

dstr = record.readframes(N * numChannels)
waveData = np.frombuffer(dstr, np.int16)
timeEMG = np.linspace(0, len(waveData)/sampleRate, num=len(waveData))

print('El registro tiene %d canal/canales' % numChannels)
print('La frecuencia de muestreo es de %d Hz.' % sampleRate)

# Graficar 0–23 segundos
t0, tf = 0, 23
t_0, t_f = int(t0*sampleRate), int(tf*sampleRate)

plt.plot(timeEMG[t_0:t_f], waveData[t_0:t_f])
plt.xlim(t0, tf)
plt.ylim(-18000, 19000)
plt.title('EMG crudo')
plt.xlabel('Tiempo [s]')
plt.ylabel('Voltaje [µV]')
plt.show()
```

## 5.3 Filtrado de la señal EMG

El preprocesamiento de EMG implica eliminar artefactos y ruidos externos para resaltar la actividad muscular real.  
Entre los filtros más comunes están:

- **Filtro pasa altas**: elimina frecuencias bajas (<20 Hz), asociadas a movimiento o desplazamiento de electrodos.  
- **Filtro pasa bajas**: elimina frecuencias muy altas (>450 Hz), generalmente ruido electrónico.  
- **Filtros paso banda**: dejan pasar solo un rango (20–450 Hz), ideal para EMG porque suprime movimiento de baja frecuencia y ruido de alta frecuencia.  
- **Filtros notch**: eliminan frecuencias específicas, como el ruido de la red eléctrica (50 o 60 Hz).  

En este ejemplo se aplica un filtro paso banda de 10–400 Hz, y luego se muestra la señal rectificada.


```python
from scipy.signal import butter, filtfilt

# Definir parámetros del filtro paso banda
nyq = sampleRate * 0.5   # frecuencia de Nyquist
low = 10 / nyq
high = 400 / nyq

# Diseñar filtro Butterworth de orden 3
b, a = butter(3, [low, high], btype='band')

# Aplicar filtro con fase cero
sen_filt = filtfilt(b, a, waveData[t_0:t_f])

# Graficar señal filtrada
plt.figure(figsize=(12,3))
plt.plot(timeEMG[t_0:t_f], sen_filt)
plt.xlabel('Tiempo [s]')
plt.ylabel('Voltaje (mV)')
plt.title(f"Señal filtrada con paso banda {t0} a {tf} segundos")
plt.show()
```

```python
# Rectificación (valores absolutos)
sen_rct = np.abs(sen_filt)

# Graficar señal rectificada
plt.figure(figsize=(12,3))
plt.plot(timeEMG[t_0:t_f], sen_rct)
plt.xlabel('Tiempo [s]')
plt.ylabel('Voltaje (mV)')
plt.title(f"Señal rectificada {t0} a {tf} segundos")
plt.show()
```

### Envolvente de la señal EMG

En este caso no usamos un filtro notch, ya que el equipo de adquisición (MuscleSpiker Box)  
ya atenúa el ruido de la red eléctrica de 50/60 Hz en el hardware.  

Por lo tanto, el flujo de procesamiento se limita a:

1. **Filtrado paso banda (10–400 Hz)** para aislar la banda de interés del EMG.  
2. **Rectificación**, que convierte la señal en valores positivos.  
3. **Envolvente**, que suaviza y muestra con claridad los periodos de activación muscular.


```python
# Envolvente con ventana de 50 ms
ventana = int(0.05 * sampleRate)
envolvente = np.convolve(sen_rct, np.ones(ventana)/ventana, mode="same")

# Graficar envolvente junto con señal rectificada
plt.figure(figsize=(12,3))
plt.plot(timeEMG[t_0:t_f], sen_rct, alpha=0.5, label="Rectificada")
plt.plot(timeEMG[t_0:t_f], envolvente, linewidth=2, label="Envolvente")
plt.xlabel('Tiempo [s]')
plt.ylabel('Voltaje (mV aprox.)')
plt.title(f"Señal EMG rectificada + envolvente {t0} a {tf} segundos")
plt.legend()
plt.show()
```

```python
from scipy.signal import hilbert

# 1. Señal cruda a float y quitar DC
emg = waveData.astype(np.float64)
emg = emg - np.mean(emg)

# 2. Filtro pasa banda 10–400 Hz
b, a = butter(3, [low, high], btype='band')
emg_filtered = filtfilt(b, a, emg)

# 3. Rectificación
emg_rect = np.abs(emg_filtered)

# 4. Envolvente con Hilbert
emg_env = np.abs(hilbert(emg_filtered))
```

```python
# 5. Validación simple del filtrado
def validar_filtro(x, y):
    min_len = min(len(x), len(y))
    x, y = x[:min_len], y[:min_len]
    corr = np.corrcoef(x, y)[0,1] if np.std(x)>0 and np.std(y)>0 else 0.0
    potencia = np.sum(np.abs(y)**2)/len(y)
    residuo = x - y
    snr = 10*np.log10(np.sum(y**2)/np.sum(residuo**2)) if np.sum(residuo**2)>0 else np.inf
    return corr, potencia, snr

corr, pot, snr = validar_filtro(emg, emg_filtered)
print(f"Correlación: {corr:.3f}, Potencia: {pot:.2f}, SNR: {snr:.2f} dB")
```

```python
# 6. Graficar crudo, filtrado, envolvente en primeros 5 s
mask = (timeEMG >= 0) & (timeEMG <= 5)

plt.figure(figsize=(12,6))
plt.plot(timeEMG[mask], emg[mask], alpha=0.3, label="EMG crudo")
plt.plot(timeEMG[mask], emg_filtered[mask], alpha=0.7, label="Filtrado (10–400 Hz)")
plt.plot(timeEMG[mask], emg_env[mask], linewidth=2, label="Envolvente (Hilbert)")
plt.title("Protocolo EMG: filtrado → rectificación → envolvente")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud (unidades del archivo)")
plt.legend()
plt.show()
```

Resultados de la validación:  

- **Correlación** ≈ 0.89 → el filtrado preserva gran parte de la forma de la señal original.  
- **Potencia** ≈ 2.3×10^5 (u.a.^2) → refleja la energía promedio de la señal filtrada.  
- **SNR** ≈ 4.7 dB → la mejora es modesta, pero suficiente para destacar periodos de activación muscular.  

En este punto se tiene un pipeline completo de EMG listo para compararse con métricas típicas de EEG.

## Comparativa EEG vs EMG con datos reales  

Para que la comparación sea más clara, aplicamos un pipeline similar en ambas señales:  

- **EEG:** filtrado en banda alfa (8–13 Hz).  
- **EMG:** filtrado 10–400 Hz + envolvente.  

Se grafican ventanas de 5 segundos para observar cómo luce cada bioseñal procesada.


```python
# --- EEG: filtrar banda alfa (8–13 Hz) ---
fs_eeg = 128
eeg_mean = df_eeg.mean(axis=1).values
nyq_eeg = fs_eeg/2
b_alpha, a_alpha = butter(3, [8/nyq_eeg, 13/nyq_eeg], btype='band')
eeg_alpha = filtfilt(b_alpha, a_alpha, eeg_mean)

t_eeg = np.arange(len(eeg_mean)) / fs_eeg

# --- EMG: ya filtrado y con envolvente (emg_env) ---
# Usamos el vector timeEMG

# Graficar 5 segundos de cada señal
tf = 5
mask_eeg = (t_eeg >= 0) & (t_eeg <= tf)
mask_emg = (timeEMG >= 0) & (timeEMG <= tf)

plt.figure(figsize=(12,6))
plt.subplot(2,1,1)
plt.plot(t_eeg[mask_eeg], eeg_alpha[mask_eeg])
plt.title("EEG filtrado en banda alfa (8–13 Hz)")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud (u.a.)")

plt.subplot(2,1,2)
plt.plot(timeEMG[mask_emg], emg_env[mask_emg])
plt.title("EMG filtrado + envolvente (10–400 Hz)")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud (u.a.)")

plt.tight_layout()
plt.show()
```

## Comparativa EEG vs EMG con filtros básicos  

Tanto el EEG como el EMG requieren filtrado inicial para eliminar artefactos y ruido.  
La diferencia central está en **qué rango interesa preservar** y en **qué significa la forma de onda resultante**.

- **EEG:** se filtra entre 0.5 y 40 Hz (Allen & Rabiner, 1977; Widmann et al., 2015). Esto elimina tendencias lentas (deriva de electrodos) y ruido de alta frecuencia.  
  El resultado conserva las oscilaciones corticales de interés:  
  - Ritmo **alfa (8–13 Hz)** → relajación, atención.  
  - Ritmo **beta (13–30 Hz)** → actividad motora, concentración.  

- **EMG:** se filtra entre 10 y 400 Hz (Merletti & Hermens, 2000; De Luca, 1997). Esto elimina el movimiento lento y el ruido eléctrico, preservando la actividad muscular rápida.  
  Posteriormente se rectifica y se calcula la envolvente, porque lo relevante no son oscilaciones, sino la magnitud y duración de la contracción.

**Clave:** el mismo método técnico (paso banda) tiene efectos y sentidos distintos dependiendo de la bioseñal (Addison, 2017).


```python
# --- EEG: filtro paso banda 0.5–40 Hz ---
nyq_eeg = fs_eeg/2
b_eeg, a_eeg = butter(3, [0.5/nyq_eeg, 40/nyq_eeg], btype='band')
eeg_band = filtfilt(b_eeg, a_eeg, eeg_mean)

# --- EMG: filtro paso banda 10–400 Hz ---
# Usamos emg_filtered (definido antes) y emg_env (envolvente Hilbert)

# Seleccionar primeros 5 segundos
tf = 5
mask_eeg = (t_eeg >= 0) & (t_eeg <= tf)
mask_emg = (timeEMG >= 0) & (timeEMG <= tf)

plt.figure(figsize=(12,8))

# EEG
plt.subplot(2,1,1)
plt.plot(t_eeg[mask_eeg], eeg_band[mask_eeg], color="tab:blue")
plt.title("EEG filtrado 0.5–40 Hz (manteniendo oscilaciones)")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud (u.a.)")

# EMG
plt.subplot(2,1,2)
plt.plot(timeEMG[mask_emg], emg_filtered[mask_emg], alpha=0.6, label="EMG filtrado (10–400 Hz)")
plt.plot(timeEMG[mask_emg], emg_env[mask_emg], linewidth=2, label="Envolvente (Hilbert)", color="tab:red")
plt.title("EMG filtrado, rectificado y con envolvente")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud (u.a.)")
plt.legend()

plt.tight_layout()
plt.show()
```

## Resumen de hallazgos y metodología 

Para cerrar la comparación, conviene sintetizar los **pipelines de procesamiento** aplicados a cada bioseñal.  
Aunque ambos parten de un filtro paso banda, los pasos siguientes divergen por completo:

- **EEG**  
  1. Filtro paso banda (0.5–40 Hz).  
  2. Análisis espectral en bandas (alfa, beta, theta, delta).  
  3. Interpretación en términos de oscilaciones corticales.  

- **EMG**  
  1. Filtro paso banda (10–400 Hz).  
  2. Rectificación de la señal.  
  3. Envolvente (promedio móvil o Hilbert).  
  4. Interpretación en términos de fuerza y duración de contracciones.

**Hallazgos principales**  

- El **mismo filtro paso banda** limpia artefactos tanto en EEG como en EMG, pero los rangos de frecuencia difieren según la fisiología.  
- En **EEG**, preservar la forma de onda es crucial para estudiar fase, frecuencia y coherencia.  
- En **EMG**, la señal rectificada + envolvente traduce la activación muscular en una curva clara y cuantificable.  
- Comparar ambas bioseñales ayuda a entender cómo las técnicas de procesamiento son universales, pero su **interpretación depende del fenómeno de origen**.
